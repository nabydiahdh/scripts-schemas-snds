```python
import re 
def find_length(x):  #fonction qui permet retourner length d'un format 
  res = re.search("\(\d+(,\d+)?\)", x)
  if res:
    return res.group().replace("(", "").replace(")", "")
  return None

##### trouver le format 
def find_format(x):
  res=x.split("(")[0]
  if res:
    return res
  return ""


df["length"]=df["format"].apply(find_length)
df["format"]=df["format"].apply(find_format)


import pandas as pd
df_tab_lib= pd.read_csv("tables_dcirs.csv", sep=",")

df_tab_lib.sample(10)
df_tab_lib.columns=["table","lib_tab"]


df["tab_lib"]=""
for i in range(len(df["tables"])):
  for j in range(len(df_tab_lib["table"])):
   if df["tables"][i]==df_tab_lib["table"][j]:
    df["tab_lib"][i]=df_tab_lib["lib_tab"][j]

df["datecreated"]="2016"
df["datedeleted"]=""
df["datemissing"]=""
df["observation"]=""
df["nomenclature"]=""

# replace NaN values with empty string 
df = df.fillna('')

```

```python
def numbint(x): 
  res = re.search(",", x)
  
  if res :
    return "number"
  return "integer"

df['type2']=df['length'].apply(numbint)
```

```python
for i in range(len(df['variables'])):
  if df['type2'][i]=='number' and df['format'][i]=='NUMBER':
    df['format'][i]='number'
  elif df['type2'][i]=='integer' and df['format'][i]=='NUMBER':
    df['format'][i]='integer'
```

```python
mapping_t = {
    "CHAR": "string",
    "NUMBER":"number",
    "DATE":"date",
    "VARCHAR2":"string",
    "INTERGER":"integer",
    "number":"number",
    "integer":"integer"
}


df["type"]=df["format"].map(mapping_t)

# fonction qui permet de passer de number to integer 

# Creation dy type oracle 
mapping_to = {
    "CHAR": "string",
    "VARCHAR2":"string",
    "DATE":"date",
    "NUMBER":"number",
    "INTEGER":"number",
    "number":"number",
    "integer":"number"
}

df["type_oracle"]=df["format"].map(mapping_to)
```

```python
from tableschema import Schema
import json
for table in tableexist['table']:
  with open('{}.json'.format(table), "r", encoding='utf-8') as f:
    descriptor = {'fields': json.load(f)}
   #descriptor="{}.json".format(table)
    schema=Schema(descriptor)
    schema.descriptor['name']=table
    schema.descriptor['title']=tableexist['tab_lib'][tableexist['table'].index(table)]
    schema.descriptor['produit']=tableexist['produit']
    schema.descriptor['missingValues'] = ['']
    if table=='NS_PRS_F': #principal table where primary key
      schema.descriptor['primaryKey']= [
        "CLE_DCI_JNT"
    ]
      schema.descriptor['foreignKeys']= [
    
        {
            "fields": "PFS_EXE_NUM",
            "reference": {
                "resource": "DA_PRA_R",
                "fields": "PFS_PFS_NUM"
            }
        },
        {
            "fields": "PFS_PRE_NUM",
            "reference": {
                "resource": "DA_PRA_R",
                "fields": "PFS_PFS_NUM"
            }
        },
        {
            "fields": "PRS_MTT_NUM",
            "reference": {
                "resource": "DA_PRA_R",
                "fields": "PFS_PFS_NUM"
            }
        },
        {
            "fields": [
                "BEN_IDT_ANO"
            ],
            "reference": {
                "resource": "IR_BEN_R",
                "fields": [
                    "BEN_IDT_ANO"
                ]
            }
        },
        {
            "fields": [
                "BEN_IDT_ANO"
            ],
            "reference": {
                "resource": "IR_IBA_R",
                "fields": [
                    "BEN_IDT_ANO"
                ]
            }
        },
        {
            "fields": [
                "ETB_PRE_FIN"
            ],
            "reference": {
                "resource": "BE_IDE_R",
                "fields": [
                    "IDE_ETA_NU8"
                ]
            }
        },
        {
            "fields": [
                "ETB_EXE_FIN"
            ],
            "reference": {
                "resource": "BE_IDE_R",
                "fields": [
                    "IDE_ETA_NU8"
                ]
            }
        }
    ]

    else :
      schema.descriptor['foreignKeys'] = [{"fields":["CLE_DCI_JNT"],
                                           "reference": {
                                             "resource": "NS_PRS_F",
                                             "fields":["CLE_DCI_JNT"]
                                             }
                                          },
                                          {"fields":["BEN_IDT_ANO"],
                                           "reference": {
                                             "resource": "IR_IBA_R",
                                             "fields":["BEN_IDT_ANO"]
                                             }
                                          }
                                          
                                         ]
    schema.descriptor['history']={"dateCreated": 2016,
                                    "dateDeleted": "",
                                    "dateMissing": []}
    schema.descriptor['champ']=""
    schema.descriptor['observation']=""
      
      
    schema.commit()
    schema.save('{}.json'.format(table),ensure_ascii=False)
```
